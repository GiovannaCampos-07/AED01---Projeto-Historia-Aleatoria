```python
import random #modulo para escolhas aleatórias

# Listas de elementos (strings) para criar as histórias.

personagens = ["uma cientista", "uma programadora", "uma princesa", "uma cozineira", "uma contadora"]
lugares = ["fazenda", "mecânica", "escola", "faculdade"]
acoes = ["desvendou uma fórmula mágica para crescer cabelo", "encontrou um mapa do tesouro que se encontra no Deserto do Atacama", "ganhou o prêmio Nobel"]

def gerar_historia():
    personagem = random.choice(personagens) # escolhe aleatoriamente na lista.
    lugar = random.choice(lugares)
    acao = random.choice(acoes)
    historia = f"{personagem} em uma {lugar} {acao}."
    return historia

print("=== Gerador de Histórias Aleatórias ===")
continuar ="sim"

while continuar.lower() == "sim":
    print("\n" + gerar_historia()) #Gera e mostra nova história.
    continuar = input("\n gerar outra história? (sim/não): ")

print("\n Obrigada por usar o gerador de histórias!")
```
